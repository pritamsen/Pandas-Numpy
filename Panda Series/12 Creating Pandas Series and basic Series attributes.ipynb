We can create a Series with a custom index and name

```python
named_series = pd.Series(data = ['Ashok', 'Kumar', 18, 183], index = ['Forename', 'Surname', 'Age', 'Height'], name = 'Student Info')
named_series
```

```python
named_series['Forename']
```

```python
named_series.dtype
```

```python
named_series.ndim
```

```python
named_series.shape
```

```python
named_series.size
```

```python
named_series.index
```

```python
named_series.values
```

```python
named_series.name
```

We can create Series from other data structures

```python
my_list = [1, 2, 3, 4]
pd.Series(my_list)
```

```python
my_tuple = (1, 2, 3, 4)
pd.Series(my_tuple)
```

```python
# my_set = {1, 2, 3, 4}
# pd.Series(my_set)  # sets are unordered and cannot be used to make a Series
```

```python
my_dict = {'A': 1, 'B': 2, 'C': 3, 'D': 4}
pd.Series(my_dict)
```

Pandas Series are strictly one dimensional

```python
pd.Series([[1, 2], [3, 4], [5, 6]])
```
